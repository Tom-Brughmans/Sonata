```python
import sys

sys.path.append("../sonata/")
import pandas as pd
from preprocessing import preprocess
```

#### Loading in dataset and site and contents information 

```python
df = pd.read_csv("../../data/SonataDataNewNew.csv")

contents = pd.read_excel("../../data/site_contents_data/amphora_content.xlsx", na_filter = False)

site = pd.read_excel(
    "../../data/site_contents_data/Template_sites_PK2.xlsx", 
    na_filter = False, 
    sheet_name = "Sheet1", 
    usecols=["Site", "NewName", "Type", "References and notes"]
)

sea_side = pd.read_excel("../../data/Template_sites_PK-1.xlsx", na_filter = False)

provenance = pd.read_excel("../../data/Tamplate_provenance-1.xlsx", na_filter = False)

type = pd.read_excel("../../data/Template_type-1.xlsx", na_filter = False)

df = df.drop(["Unnamed: 0"], axis=1)
```

```python
df["Site"] = df["Site"].replace({"Romanità di Cattolica ": "Romanità di Cattolica",
                                  "Pyrgi": "Caere-Pyrgi"})
                                
df["Provenance"] = df["Provenance"].replace({"Baetice": "Baetica"})
df
```

```python
site = site[site.Site != "Pyrgi"]

site = site.rename(columns = {
                       "NewName": "Grouped_sites",
                       "Type":"Site_type", 
                       "References and notes":"Site_notes"
                       })

site.head()
```

```python
contents = contents.rename(columns = {
                               "Amphora type": "amphora_type", 
                               "notes": "Amphora_type_notes"
                               })
contents.head()
```

### MERGING SITE AND DF

```python
df_merged = pd.merge(df, site, on = "Site")
df_merged
```

### MERGING CONTENTS TO THE OTHER TWO

```python
df_merged["Amphora_type_main"] = df_merged["Amphora_type"].str.split("_", n = 1, expand = True)[0]
```

```python
df_merged["Amphora_type_main"] = df_merged["Amphora_type_main"].replace({"Dr 2-5 ": "Dr 2-5", 
                                                                         "Dr 7-11 ": "Dr 7-11", 
                                                                         "Garloise 1": "Gauloise 1", 
                                                                         "Dr 20 ":"Dr 20", 
                                                                         "Ostia LIX/XXIII": "Ostia XXIII/LIX"
                                                                         })
```

```python
df_merged1 = df_merged.merge(contents, how = "left", right_on = "amphora_type", left_on = "Amphora_type_main")
df_merged1.rename(columns = {"Lower_ context_date":"Lower_context_date"}, inplace = True)
df_merged1
```

# ADD 'Tyrrhenian vs Adriatic' column

```python
# check whether there any sites that are missing or misspelled

df_sites = list(df_merged["Site"].unique())
sea_side = sea_side[sea_side.Site != "Pyrgi"]
sea_side = sea_side.rename(columns = {"Tyrrhenian vs Adriatic": "Tyrrhenian_vs_Adriatic"})

sea_side_sites = list(sea_side["Site"].unique())
difference = list(set(sea_side_sites) - set(df_sites))
difference
```

```python
sea_side["Site"] = sea_side["Site"].replace({"Aquae Statiellae,  VIA AURELIANO\nGALEAZZO, corso Cavour": "Aquae Statiellae,  VIA AURELIANO GALEAZZO, corso Cavour",
                                             "Scopietto, Umbria": "Scoppieto, Umbria", 
                                             "Romanità di Cattolica\n": "Romanità di Cattolica"})
                                             
sea_side = sea_side[["Site", "Tyrrhenian_vs_Adriatic"]]
```

```python
df_merged1 = pd.merge(df_merged1, sea_side,  on = "Site")
df_merged1
```

# CORRECT PROVENANCES

```python
provenance = provenance[provenance["New Provenance"] != "" ]
old_prov, new_prov = provenance["Provenance"].tolist(), provenance["New Provenance"].tolist()
```

```python
for old, new in zip(old_prov, new_prov):
    df_merged1["Provenance"] = df_merged1["Provenance"].replace({old: new})
    
df_merged1["Provenance"].unique()
```

# ADD GROUPED TYPE NAMES 

```python
type.rename(columns = {type.columns[1]: "New_type_name", "Amphora_type": "Type"}, inplace = True)
type 
```

```python
df_merged1 = pd.merge(df_merged1, type, how = "left", right_on = "Type", left_on = "Amphora_type")
df_merged1
```

```python
# merge New_type_name with values from Amphora type
df_merged1['Amphora_type_merged'] = df_merged1.apply(lambda row: row['Amphora_type'] if row['New_type_name'] == '' else row['New_type_name'], axis=1)
```

# Preprocess columns 

```python
df_merged1.columns.unique()
```

```python
df_merged1 = df_merged1.drop(["Amphora_type_main", "amphora_type", "Type"], axis=1)
df_merged1
```

```python
cols_for_preprocessing = [
    "Site",
    "Site_type",
    "Provenance", 
    "Amphora_type",
    "Amphora_type_merged",
    "content",
    "Grouped_sites",
    "Tyrrhenian_vs_Adriatic", 
    "New_type_name"]

for column in cols_for_preprocessing:
    df_merged1[column] = preprocess(df_merged1[column])

df_merged1
```

```python
colms_to_numeric = ["Amphora_type_upper_date", "Amphora_type_lower_date",  "Lower_context_date", "Upper_context_date"]


for column in colms_to_numeric:
    df_merged1[column] = pd.to_numeric(df_merged1[column], errors='coerce') 
df_merged1
```

```python
df_merged1.to_csv('../../data/Sonata_data.csv')
```
