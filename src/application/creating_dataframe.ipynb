```python
import sys

sys.path.append("../sonata/")
import pandas as pd
from preprocessing import preprocess
```

#### Loading in dataset and site and contents information 

```python
df = pd.read_csv("../../data/SonataDataNewNew.csv")

contents = pd.read_excel("../../data/site_contents_data/amphora_content.xlsx", na_filter = False)

site = pd.read_excel(
    "../../data/site_contents_data/Template_sites_PK2.xlsx", 
    na_filter = False, 
    sheet_name = "Sheet1", 
    usecols=["Site", "NewName", "Type", "References and notes"]
)

df = df.drop(["Unnamed: 0"], axis=1)
df
```

```python
df["Site"] = df["Site"].replace({"Romanità di Cattolica ": "Romanità di Cattolica",
                                  "Pyrgi": "Caere-Pyrgi"})
                                
df["Provenance"] = df["Provenance"].replace({"Baetice": "Baetica"})
df
```

```python
site = site[site.Site != "Pyrgi"]

site = site.rename(columns = {
                       "NewName": "Grouped_sites",
                       "Type":"Site_type", 
                       "Site": "site",
                       "References and notes":"Site_notes"
                       })

site.head()
```

```python
contents = contents.rename(columns = {
                               "Amphora type": "amphora_type", 
                               "notes": "Amphora_type_notes"
                               })
contents.head()
```

### MERGING SITE AND DF

```python
df_merged = pd.merge(df, site, how = "left", right_on = "site", left_on = "Site")
df_merged
```

### MERGING CONTENTS TO THE OTHER TWO

```python
df_merged["Amphora_type_main"] = df_merged["Amphora_type"].str.split("_", n = 1, expand = True)[0]
```

```python
df_merged["Amphora_type_main"] = df_merged["Amphora_type_main"].replace({
                                                                    "Dr 2-5 ": "Dr 2-5", 
                                                                    "Dr 7-11 ": "Dr 7-11", 
                                                                    "Garloise 1": "Gauloise 1", 
                                                                    "Dr 20 ":"Dr 20", 
                                                                    "Ostia LIX/XXIII": "Ostia XXIII/LIX"
                                                                    })
```

```python
df_merged1 = df_merged.merge(contents, how = "left", right_on = "amphora_type", left_on = "Amphora_type_main")
df_merged1.rename(columns = {"Lower_ context_date":"Lower_context_date"}, inplace = True)
df_merged1
```

```python
cols_for_preprocessing = [
    "Site",
    "site", 
    "Provenance",
    "amphora_type", 
    "Amphora_type",
    "content",
    "Amphora_type_main",
    "Grouped_sites"]

for column in cols_for_preprocessing:
    df_merged1[column] = preprocess(df_merged1[column])

df_merged1
```

```python
colms_to_numeric = ["Amphora_type_upper_date", "Amphora_type_lower_date",  "Lower_context_date", "Upper_context_date"]


for column in colms_to_numeric:
    df_merged1[column] = pd.to_numeric(df_merged1[column], errors='coerce') 
df_merged1
```

```python
df_merged1.to_csv('../../data/Sonata_data_site_contents.csv')
```
